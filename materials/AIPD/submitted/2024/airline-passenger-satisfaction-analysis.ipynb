**In almost all the columns which start with "SAT_" (satisfaction rating) there is at least one row, where the value is 0. I don't want to use these rows, because the 0 could be interpreted as a worse rating than a 1, but it is not true.**

**I need to check how many rows are there which are like this, and if there isn't a lot, I'll remove those lines from the df.**

```python
# prompt: How many rows are there in the df, where in either of the columns that start with "SAT_" there is at least one 0 value

import pandas as pd

# Assuming df is already loaded as in the provided code

sat_cols = [col for col in df.columns if col.startswith('SAT_')]
rows_with_zeros = df[(df[sat_cols] == 0).any(axis=1)]
num_rows_with_zeros = len(rows_with_zeros)

print(f"Number of rows with at least one 0 in SAT_ columns: {num_rows_with_zeros}")
```

```python
# prompt: remove every row from df, where in either of the columns that start with "SAT_" there is at least one 0 value

# Filter out rows where any column starting with "SAT_" contains a 0
sat_cols = [col for col in df.columns if col.startswith('SAT_')]
df = df[~(df[sat_cols] == 0).any(axis=1)]
```

```python
# prompt: how many different vlaues do the columns have, and what type they are? if they are not numerical, write out all the possible values of them also

import pandas as pd

# Assuming df is already loaded and processed as in the provided code
# ... (previous code to load and clean the dataframe)

# Now, analyze column values and types
for col in df.columns:
    print(f"Column: {col}")
    print(f"  Unique Values: {df[col].nunique()}")
    print(f"  Data Type: {df[col].dtype}")

    if not pd.api.types.is_numeric_dtype(df[col]):
        print(f"  Possible Values: {df[col].unique()}")
    print("-" * 20)
```

**The non-numerical columns all have either 2 or 3 unique values. I will one-hot-encode them**

```python
# prompt: one-hot-encode the columns that are non-numerical
# (use 0s and 1s as the new values, not boolean type values).

# Assuming df is already loaded and processed as in the provided code
# ... (previous code to load and clean the dataframe)


def one_hot_encode(df, column):
    values = df[column].unique()
    for value in values:
        new_col_name = f"{column}_{value}"
        df[new_col_name] = (df[column] == value).astype(int)
    df = df.drop(columns=[column])
    return df

for col in df.columns:
    if not pd.api.types.is_numeric_dtype(df[col]):
        df = one_hot_encode(df, col)

df.head()
```

**I want to see which features correlate with which features the most**

```python
# prompt: make a correlation matrix from df

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already loaded and processed as in the previous code

# Calculate the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
# print(corr_matrix)

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Airline Passenger Satisfaction')
plt.show()
```

```python
# prompt: print the different columns' correlations with the "Airline Satisfaction_satisfied" column

# Assuming df is already loaded and processed as in the previous code

# Print correlations with 'Airline Satisfaction_satisfied'
print(df.corr()['Airline Satisfaction_satisfied'].sort_values(ascending=False))
```

```python
# prompt: I want to see a bar chart: on the x axis all the features which start with "Class_",
# on the y axis the number of rows belonging  to that class, plus separate the columns into
#2 columns in each Class type, based on if their "Airline Satisfaction_satisfied" or
#their "Airline Satisfaction_neutral or dissatisfied" value is 1.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already loaded and processed as in the provided code
# ... (previous code to load and clean the dataframe)

# Filter columns starting with "Class_"
class_columns = [col for col in df.columns if col.startswith("Class_")]

# Create the bar chart
plt.figure(figsize=(10, 6))

for class_col in class_columns:
    satisfied_count = df[df['Airline Satisfaction_satisfied'] == 1][class_col].sum()
    neutral_dissatisfied_count = df[df['Airline Satisfaction_neutral or dissatisfied'] == 1][class_col].sum()

    plt.bar(class_col + " - Satisfied", satisfied_count, label='Satisfied', color='limegreen')
    plt.bar(class_col + " - Neutral/Dissatisfied", neutral_dissatisfied_count, label='Neutral or Dissatisfied', color='tomato')

plt.xlabel("Class Type")
plt.ylabel("Number of Passengers")
plt.title("Passenger Count by Class Type")

plt.xlabel("Class Type and Satisfaction")
plt.ylabel("Number of Passengers")
plt.title("Passenger Count by Class and Satisfaction Level")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Filter columns starting with "Class_"
class_columns = [col for col in df.columns if col.startswith("Type of Travel_")]

# Create the bar chart
plt.figure(figsize=(9, 7))

for class_col in class_columns:
    satisfied_count = df[df['Airline Satisfaction_satisfied'] == 1][class_col].sum()
    neutral_dissatisfied_count = df[df['Airline Satisfaction_neutral or dissatisfied'] == 1][class_col].sum()

    plt.bar(class_col + " - Satisfied", satisfied_count, label='Satisfied', color='limegreen')
    plt.bar(class_col + " - Neutral/Dissatisfied", neutral_dissatisfied_count, label='Neutral or Dissatisfied', color='tomato')

plt.xlabel("Type of Travel")
plt.ylabel("Number of Passengers")
plt.title("Passenger Count by Type of Travel")

plt.xlabel("Type of Travel and Satisfaction")
plt.ylabel("Number of Passengers")
plt.title("Passenger Count by Type of Travel and Satisfaction Level")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
