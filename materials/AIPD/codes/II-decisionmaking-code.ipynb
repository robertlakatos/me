```python
# prompt: Modify the code and apply the Moving Averages, Relative Strength Index (RSI), Bollinger Bands for the data. Present the result in the dataframe.

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the list of tickers
tickers = ["AAPL", "MSFT", "GOOG", "AMZN", "META", "INTC", "AMD", "NVDA", "TSM",
           "ADBE", "CRM", "ORCL", "SAP"]

# Define the date range
start_date = "2023-01-01"
end_date = "2023-12-31"

# Download data for each ticker and store it in a dictionary
stock_data = {}
for ticker in tickers:
  data = yf.download(ticker, start=start_date, end=end_date)
  stock_data[ticker] = data

# Function to calculate moving averages
def calculate_moving_averages(df, window):
  df['MA_' + str(window)] = df['Close'].rolling(window=window).mean()
  return df

# Function to calculate RSI
def calculate_rsi(df, window):
  delta = df['Close'].diff()
  gain = (delta.where(delta > 0, 0)).fillna(0)
  loss = (-delta.where(delta < 0, 0)).fillna(0)
  avg_gain = gain.rolling(window=window).mean()
  avg_loss = loss.rolling(window=window).mean()
  rs = avg_gain / avg_loss
  rsi = 100 - (100 / (1 + rs))
  df['RSI_' + str(window)] = rsi
  return df

# Function to calculate Bollinger Bands
def calculate_bollinger_bands(df, window):
  rolling_mean = df['Close'].rolling(window=window).mean()
  rolling_std = df['Close'].rolling(window=window).std()
  df['BB_Upper_' + str(window)] = rolling_mean + (rolling_std * 2)
  df['BB_Lower_' + str(window)] = rolling_mean - (rolling_std * 2)
  return df

# Calculate indicators for each stock
for ticker, data in stock_data.items():
  data = calculate_moving_averages(data, 20)
  data = calculate_rsi(data, 14)
  data = calculate_bollinger_bands(data, 20)
  stock_data[ticker] = data

# Display the data with indicators for a specific stock (e.g., AAPL)
print(stock_data['AAPL'])
```

```python
# prompt: modify the that i can see all stock's RSI index in one table.

import pandas as pd
# Define the list of tickers
tickers = ["AAPL", "MSFT", "GOOG", "AMZN", "META", "INTC", "AMD", "NVDA", "TSM",
           "ADBE", "CRM", "ORCL", "SAP"]

# Define the date range
start_date = "2023-01-01"
end_date = "2023-12-31"

# Download data for each ticker and store it in a dictionary
stock_data = {}
for ticker in tickers:
  data = yf.download(ticker, start=start_date, end=end_date)
  stock_data[ticker] = data

# Function to calculate RSI
def calculate_rsi(df, window):
  delta = df['Close'].diff()
  gain = (delta.where(delta > 0, 0)).fillna(0)
  loss = (-delta.where(delta < 0, 0)).fillna(0)
  avg_gain = gain.rolling(window=window).mean()
  avg_loss = loss.rolling(window=window).mean()
  rs = avg_gain / avg_loss
  rsi = 100 - (100 / (1 + rs))
  df['RSI_' + str(window)] = rsi
  return df

# Calculate RSI for each stock and store it in the DataFrame
for ticker, data in stock_data.items():
  stock_data[ticker] = calculate_rsi(data, 14)

# Create a DataFrame to store the latest RSI values for all stocks
rsi_table = pd.DataFrame(columns=['Ticker', 'Latest RSI'])

for ticker, data in stock_data.items():
  latest_rsi = data['RSI_14'].iloc[-1]  # Get the last RSI value
  rsi_table = pd.concat([rsi_table, pd.DataFrame({'Ticker': [ticker], 'Latest RSI': [latest_rsi]})], ignore_index=True)

# Display the RSI table
rsi_table
```
